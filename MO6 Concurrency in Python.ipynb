```python
from datetime import date

# Get the current date
today = date.today()

# Convert to string
today_string = str(today)

# Write to file
with open('today.txt', 'w') as file:
    file.write(today_string)

print(f"Current date '{today_string}' written to today.txt")
```

```python
# Exercise 13.2: Read today.txt into the string today_string

# Read the file
with open('today.txt', 'r') as file:
    today_string = file.read()

print(f"Contents of today.txt: {today_string}")
print(f"Type: {type(today_string)}")
```

```python
# Exercise 13.3: Parse the date from today_string
from datetime import datetime

# Parse the date string (assuming ISO format YYYY-MM-DD)
parsed_date = datetime.strptime(today_string, '%Y-%m-%d').date()

print(f"Original string: {today_string}")
print(f"Parsed date object: {parsed_date}")
print(f"Type: {type(parsed_date)}")
print(f"\nDate components:")
print(f"  Year: {parsed_date.year}")
print(f"  Month: {parsed_date.month}")
print(f"  Day: {parsed_date.day}")
```

```python
# Exercise 15.1: Create three processes with random wait times
import multiprocessing
import time
import random
from datetime import datetime

def wait_and_print(process_num):
    """
    Function to be executed by each process.
    Waits a random time between 0 and 1 second, then prints the current time.
    """
    # Generate random wait time between 0 and 1 second
    wait_time = random.random()
    
    # Wait for the random amount of time
    time.sleep(wait_time)
    
    # Get and print the current time
    current_time = datetime.now().strftime('%Y-%m-%d %H:%M:%S.%f')
    print(f"Process {process_num}: Waited {wait_time:.4f} seconds. Current time: {current_time}")

if __name__ == '__main__':
    # Create a list to hold our processes
    processes = []
    
    print("Creating and starting three processes...\n")
    
    # Create and start three processes
    for i in range(3):
        process = multiprocessing.Process(target=wait_and_print, args=(i+1,))
        processes.append(process)
        process.start()
    
    # Wait for all processes to complete
    for process in processes:
        process.join()
    
    print("\nAll processes have completed.")
```
